### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv
# 開啟 CSV 檔案
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    #for row in rows:
        #print(row)
```

```python
# 1. 取出班次一的每一個時間
result=[]
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    Dict = csv.DictReader(csvfile)
    for row in Dict:
        result.append(row['班次1'])
    print(result)
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

import pandas as pd
raw_data = {'班次1': result}
df = pd.DataFrame(raw_data, columns = ['班次1'])
df
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存
import pandas as pd
result = [[] for i in range(5)]
with open('./data/example.csv', newline='',encoding="utf-8") as csvfile:
    Dict = csv.DictReader(csvfile)
    for row in Dict:
        result[0].append(row['班次1'])
        result[1].append(row['班次2'])
        result[2].append(row['班次3'])
        result[3].append(row['班次4'])
        result[4].append(row['班次5'])

raw_data = {'班次1': result[0], 
        '班次2': result[1], 
        '班次3': result[2], 
        '班次4': result[3],
        '班次5': result[4]}
df = pd.DataFrame(raw_data, columns = ['班次1', '班次2', '班次3', '班次4', '班次5'])
df.to_csv('save.csv')
df
```
