### 1. Dcard ç¶²å€ï¼š https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request å–å›ä¹‹å¾Œè©²æ€éº¼å–å‡ºè³‡æ–™ï¼Œè³‡æ–™å‹æ…‹æ˜¯ä»€éº¼ï¼Ÿ =>')

response = r.text
#print(response[0:3000])
print(type(response))

```

```python
print('ç‚ºä»€éº¼è¦ä½¿ç”¨ BeatifulSoup è™•ç†ï¼Ÿè™•ç†å¾Œçš„å‹æ…‹æ˜¯ä»€éº¼ï¼Ÿ => ')

soup = BeautifulSoup(response, "html5lib")#lxml
#print(soup)
print(type(soup))
```

```python
p=soup.find_all('a', class_='PostEntry_root_V6g0rd')
print(type(p))

for t in p:
    ti=t.find('h3', class_='Title__Text-v196i6-0 gmfDU')
    if(ti):
        print(ti.text)
```

### 3. è§€å¯Ÿä¸€ä¸‹çŸ¥ä¹å›ä¾†çš„è³‡æ–™å¥½åƒæœ‰é»æ€ªæ€ªçš„ï¼Œè©²æ€éº¼è§£æ±ºï¼Ÿ

```python
import requests
url = 'https://www.zhihu.com/explore'
header = {'user-agent': 'test'}
r = requests.get(url,headers=header)

r.encoding = 'utf-8'
print(r.text[0:600])
```

```python
response=r.text
soup = BeautifulSoup(response, "html5lib")
#print(soup.prettify())
titles=soup.find_all('a', class_='ExploreSpecialCard-title')
for t in titles:
    print(t.text)
```
