## ä½œæ¥­ç›®æ¨™

* è«‹åˆ©ç”¨ API: https://www.dcard.tw/_api/forums/pet/posts?popular=true å›ç­”ä¸‹åˆ—å•é¡Œï¼š

1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import json
import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text
data=json.loads(response)
num=len(data)
print(num)

for i in data[0]:
    print(i)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for i in range(num):
    print("æ¨™é¡Œ:{} è²¼æ–‡æ™‚é–“:{} ç•™è¨€äººæ•¸:{} æŒ‰è®šäººæ•¸:{} ".format(data[i]['title'],data[i]['createdAt'],data[i]['commentCount'],data[i]['likeCount']))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text
pop=json.loads(response)
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response=r.text
non_pop=json.loads(response)

popcomment=0
poplike=0
popnum=0
non_popcomment=0
non_poplike=0
non_popnum=0
for i in pop:
    popcomment=popcomment+i['commentCount']
    poplike=poplike+i['likeCount']
    popnum=popnum+1

for i in non_pop:
    non_popcomment=non_popcomment+i['commentCount']
    non_poplike=non_poplike+i['likeCount']
    non_popnum=non_popnum+1   

print("ç†±é–€ ç•™è¨€äººæ•¸{}/æŒ‰è®šäººæ•¸{}".format(popcomment/popnum,poplike/popnum))
print("éç†±é–€ ç•™è¨€äººæ•¸{}/æŒ‰è®šäººæ•¸{}".format(non_popcomment/non_popnum,non_poplike/non_popnum))
```
